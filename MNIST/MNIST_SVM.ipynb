### Using SupportVectorMachine algorithm for digit prediction on MNIST data

The data set used is a simplified version of the original MNIST, and it's divided in 2 dataset: mnist_train and mnist_test.

```python
#! pip install scikit-learn
#! pip install pandas
#! pip install numpy
```

```python
import pandas as pd

# Load training and testing data
train_data = pd.read_csv('mnist_train.csv')
test_data = pd.read_csv('mnist_test.csv')
train_data.head()
```

```python
train_data.describe()
```

```python
test_data.describe()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Preprocess the data
X_train = train_data.drop(columns=['label'])
y_train = train_data['label']
X_test = test_data.drop(columns=['label'])
y_test = test_data['label']

X_train
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

# Train the Support Vector Machine model
svm_model = SVC(kernel='rbf', gamma='scale')  # You can adjust parameters like kernel type and regularization
svm_model.fit(X_train_scaled, y_train)
```

```python
# Display the parameters of the SVM model
print("Support Vector Machine (SVM) Parameters:")
#print("Coefficients (weights):", svm_model.coef_) if using a linear kernel
print("Intercept (bias):", svm_model.intercept_)
print("Support vectors:", svm_model.support_vectors_)
print("Number of support vectors for each class:", svm_model.n_support_)
print("Support vector indices:", svm_model.support_)
print("Dual coefficients:", svm_model.dual_coef_)
print("Classes:", svm_model.classes_)
print("Kernel:", svm_model.kernel)
print("Gamma:", svm_model.gamma)
print("Regularization parameter (C):", svm_model.C)
```

```python
# Predict labels for the testing data
y_pred_svm = svm_model.predict(X_test_scaled)

# Calculate accuracy
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("Accuracy (SVM):", accuracy_svm)
```

```python
# Select the features of the first row of the training data
first_row_features = X_train_scaled[[0]]

# Predict the label using the trained logistic regression model
predicted_label = svm_model.predict(first_row_features)

# Print the predicted label
print("Predicted Label:", predicted_label)
```
